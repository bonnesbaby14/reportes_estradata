```python
import pandas as pd

parciales=pd.read_excel("./ACUMULADO.xlsx")

finales=dict()


parciales.reset_index()


for index, row in parciales.iterrows():
  if index==0:
    valores=row.keys().to_list()
    valores.append("VISITAS")
    result=pd.DataFrame(columns=valores)
 
  if str(row["Resultado visita"])=="nan" :
    continue 

  if(int (row["FOLIO"]) in finales.keys()):
    if finales[int(row["FOLIO"])]["ELEMENTO"][15]=="LOCALIZADA":
      if row["INICIO"]+row["TERMINO"] in finales[int(row["FOLIO"])]["ULTIMA"]:
        pass
      else:
        if row["Resultado visita"]=="LOCALIZADA":
          if finales[int(row["FOLIO"])]["ELEMENTO"][12]<row["TERMINO"]:
            finales[int(row["FOLIO"])]["ELEMENTO"]=row.to_list()
        finales[int(row["FOLIO"])]["CANTIDAD"]+=1
        finales[int(row["FOLIO"])]["ULTIMA"].append(row["INICIO"]+row["TERMINO"])
    else:
      if row["INICIO"]+row["TERMINO"] in finales[int(row["FOLIO"])]["ULTIMA"]:
        pass
      else:
        finales[int(row["FOLIO"])]["CANTIDAD"]+=1
        finales[int(row["FOLIO"])]["ULTIMA"].append(row["INICIO"]+row["TERMINO"])
      if row["Resultado visita"]=="LOCALIZADA":
        finales[int(row["FOLIO"])]["ELEMENTO"]=row.to_list()    
      elif finales[int(row["FOLIO"])]["ELEMENTO"][12]<row["TERMINO"]:
        finales[int(row["FOLIO"])]["ELEMENTO"]=row.to_list()
  else:
    finales[int(row["FOLIO"])]={"ELEMENTO":row.to_list(),"ULTIMA":[row["INICIO"]+row["TERMINO"]],"CANTIDAD":1}



contador=0
#print(result);

for clave in finales:
  datos=finales[clave]["ELEMENTO"]
  datos.append(finales[clave]["CANTIDAD"])
  result.loc[contador]=datos
  contador+=1

result.to_excel("./reportefinal.xlsx")

```
